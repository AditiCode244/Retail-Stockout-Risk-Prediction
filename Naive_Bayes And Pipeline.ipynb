# Naive Bayes Classification Project

## Stockout Risk Prediction for Retail Inventory

**Company:** Walmart (Retail Industry)


## Problem Faced By Company
Walmart and other similar retail companies  have to manage stock properly. Sometimes products go out of stock when customer demand suddenly increases or supplier delivery gets delayed. This creates a stockout problem, and customers may not get the product on time.

## Solution For The Problem
To solve this problem, we use **Naive Bayes classification model**.  
This model will ppredicts whether whether there is a stockout risk or not.
The model learns from inventory related data like current stock, daily demand, lead time, supplier reliability, and promotion status.

## Purpose of this Project

The purpose of this project is to  predict stockout risk in a retail store using the Naive Bayes algorithm.

## Import Libraries

```python
import pandas as pd                   
import numpy as np
```

## Load the Dataset
Here we load the dataset and preview the first few rows.

```python
df = pd.read_csv("Dataset/inventory_demand_stockout_risk.csv")
df.head()       # Load the dataset 
```

```python
df.info()  # Information About The Dataset 
```

```python
df.columns # Check the columns name
```

```python
df["stockout_risk"].value_counts()  # Count the target count of stockout risk
```

##  Selecting feature and target
- **X (Feature):** Features that are used for prediction 
- **y (Target):** Stockout Risk

I removed some columns from the features:
- stockout_risk: because it is target column.
- product_id : because it is only a identification column.
- date: because it is not directly used in this model.
- store_id: because it is also an identification column.
- weather_impact: because I am not using this column in this model.

So, the remaining columns are used as input features to predict stockout_risk.

```python
# Select features and target
X = df.drop(["stockout_risk", "product_id", "date", "store_id","weather_impact"], axis = 1)
y = df["stockout_risk"] 
```

```python
X.head() # Features
```

```python
y.head() # Target
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

In this dataset, columns like **promotion_active** and **stockout_risk** contain categorical values, so I converted them into numbers before splitting the data.

```python
X["promotion_active"] = LabelEncoder().fit_transform(X["promotion_active"])
y = le.fit_transform(y)
```

##  Import Train-Test Split

```python
from sklearn.model_selection import train_test_split 
```

```python
# Split the data into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

## Train Model Using Naive Bayes
- Here I trained the model using the Multinomial Naive Bayes algorithm.
- Naive Bayes is a classification algorithm that predicts the class based on probability.
- In this project, it predicts the stockout risk category.

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

##  Making Predictions
After training the model, I used the testing data to make predictions.        
The model predicts whether stockout risk is present or not for the test records.

```python
y_pred = model.predict(X_test)
y_pred[:10] # viewing the top 10 predicted values
```

```python
y_test[:10]  # viewing the top 10 actual values
```

##  Checking Model Score
Here I checked the accuracy score of the model which tells how many predictions are correct out of total predictions.

```python
model.score(X_test, y_test)
```

##  Evaluation
We evaluate the model using Classification Report

```python
from sklearn.metrics import classification_report
```

Classification report shows:

- **Precision:** How many predicted values were actually correct
- **Recall:** How many actual values were correctly found by the model
- **F1-score:** Balance between precision and recall

```python
report = classification_report(y_test, y_pred)
print(report)     # Model Evaluation
```

## Conclusion

- In this project, we used the Naive Bayes classification model to predict stockout risk in retail inventory.
- The model learned from inventory-related features and predicted whether there is a chance of stockout or not which help retail companies like Walmart to manage stock properly and avoid product unavailability.

Overall, this project shows that stockout risk prediction can support better decision-making in retail sales and inventory planning.

# Pipeline Model Using Naive Bayes

Now I am applying the same Naive Bayes model using Pipeline.    
In normal Naive Bayes code, model training and prediction are written separately. But in pipeline we arrange the model in one proper structure.

```python
# Importing Pipeline
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
```

```python
pipe = Pipeline([
    ("nb", MultinomialNB())
])
pipe.fit(X_train, y_train)
```

```python
y_pred_pipe = pipe.predict(X_test)
y_pred_pipe[:10]   # predicting values
```

```python
y_test[:10] # actual values
```

```python
pipe.score(X_test, y_test)
```
